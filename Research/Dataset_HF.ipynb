```python
import os
from tqdm.auto import tqdm
from pathlib import Path
from concurrent.futures import ThreadPoolExecutor, as_completed

# Setup folders
output_path = Path('/content/arxiv_data')
for s in ['train', 'val', 'test']:
    os.makedirs(output_path / f'images/{s}', exist_ok=True)
    os.makedirs(output_path / f'labels/{s}', exist_ok=True)

print("Starting ULTRA-SAFE Extraction with Live Progress (Workers: 6 | Batch: 75)...")

targets = {
    'train': 6864,
    'val': 640,
    'test': 496
}

# 1. The Shielded Worker (Resume-Capable)
def save_file_worker(img, bboxes, cat_ids, split, f_name):
    img_file = output_path / f'images/{split}' / f"{f_name}.png"
    lbl_file = output_path / f'labels/{split}' / f"{f_name}.txt"

    # SAFETY: If file exists, skip it! (Allows instant resuming if paused/crashed)
    if os.path.exists(img_file) and os.path.exists(lbl_file):
        return True

    try:
        # Save Image
        img.save(img_file)

        # Save YOLO Label
        img_w, img_h = img.size
        with open(lbl_file, 'w') as f:
            for bbox, cat_id in zip(bboxes, cat_ids):
                x, y, w, h = bbox
                # Convert to YOLO format
                f.write(f"{cat_id-1} {(x+w/2)/img_w:.6f} {(y+h/2)/img_h:.6f} {w/img_w:.6f} {h/img_h:.6f}\n")
        return True
    except Exception as e:
        return f"Error on {f_name}: {e}"

# 2. Batched Execution Logic
for split in ['train', 'val', 'test']:
    count = 0
    target_limit = targets[split]
    batch_size = 75 # UPDATED: Perfectly balanced RAM safety

    # Enhanced Live UI for Colab
    print(f"\n--- Processing {split.upper()} Split ---")
    scan_pbar = tqdm(total=target_limit, desc=f"Scanning", position=0, leave=True, dynamic_ncols=True)
    save_pbar = tqdm(total=target_limit, desc=f"💾Saving  ", position=1, leave=True, dynamic_ncols=True)

    # UPDATED: 6 Workers for a fast but stable I/O flow
    with ThreadPoolExecutor(max_workers=6) as executor:
        futures = []

        for idx, row in enumerate(dataset[split]):
            if count >= target_limit:
                break

            meta = row.get('metadata', {})

            if meta.get('doc_category') == 'scientific_articles':
                img = row['image']
                bboxes = row['bboxes']
                cat_ids = row['category_id']

                doc_id = meta.get('doc_hash') or meta.get('doc_filename') or f"doc_{idx}"
                page_num = meta.get('page_no', f"p{idx}")
                f_name = f"{doc_id}_{page_num}_{count}"

                futures.append(executor.submit(save_file_worker, img, bboxes, cat_ids, split, f_name))
                count += 1
                scan_pbar.update(1)

                # Process batch when it hits 75
                if len(futures) >= batch_size:
                    for future in as_completed(futures):
                        result = future.result()
                        if result is not True:
                            tqdm.write(f"{result}")
                        save_pbar.update(1)
                    futures = []

        # Process the final leftover files
        for future in as_completed(futures):
            result = future.result()
            if result is not True:
                tqdm.write(f"{result}")
            save_pbar.update(1)

    scan_pbar.close()
    save_pbar.close()

print("\nEXTRACTION COMPLETE! (Fully verified and safe)")

# Final Verification Count
print("\n--- FINAL FILE COUNTS ---")
for s in ['train', 'val', 'test']:
    img_count = len(os.listdir(output_path / f'images/{s}'))
    lbl_count = len(os.listdir(output_path / f'labels/{s}'))
    print(f"{s.capitalize()}: {img_count} Images, {lbl_count} Labels")
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuration
labels_path = '/content/arxiv_data/labels'
class_names = [
    "Caption", "Footnote", "Formula", "List-item", "Page-footer",
    "Page-header", "Picture", "Section-header", "Table", "Text", "Title"
]

def get_label_stats(split_name=None):
    stats = {name: 0 for name in class_names}

    # Determine which folders to scan
    folders = [split_name] if split_name else ['train', 'val', 'test']

    for folder in folders:
        path = os.path.join(labels_path, folder)
        if not os.path.exists(path): continue
        for label_file in os.listdir(path):
            with open(os.path.join(path, label_file), 'r') as f:
                for line in f:
                    try:
                        class_id = int(line.split()[0])
                        stats[class_names[class_id]] += 1
                    except: continue
    return stats

# --- 1. COLLECT ALL DATA ---
all_stats = get_label_stats()  # Total
train_stats = get_label_stats('train')
val_stats = get_label_stats('val')
test_stats = get_label_stats('test')

# Create a combined DataFrame for the report
df_report = pd.DataFrame({
    'Class': class_names,
    'TRAIN': [train_stats[c] for c in class_names],
    'VAL': [val_stats[c] for c in class_names],
    'TEST': [test_stats[c] for c in class_names],
    'TOTAL': [all_stats[c] for c in class_names]
})

# --- 2. PRINT TEXT SUMMARY ---
print("📋 --- DETAILED LABEL COUNT REPORT --- 📋")
print(df_report.to_string(index=False))
print(f"\n🔥 GRAND TOTAL BOXES: {df_report['TOTAL'].sum():,}")

# --- 3. VISUALIZATION ---
fig = plt.figure(figsize=(20, 15))
plt.suptitle("ArXiv Dataset Label Distribution Analysis", fontsize=22, fontweight='bold')

# Top Row: Grand Total Charts
ax1 = plt.subplot2grid((2, 2), (0, 0))
ax2 = plt.subplot2grid((2, 2), (0, 1))

# Bar Chart (Total)
sns.barplot(x='TOTAL', y='Class', data=df_report.sort_values('TOTAL', ascending=False), ax=ax1, palette='viridis')
ax1.set_title("Grand Total Counts (All Splits)", fontsize=16)

# Pie Chart (Total %)
ax2.pie(df_report['TOTAL'], labels=df_report['Class'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', 11))
ax2.set_title("Overall Percentage Breakdown", fontsize=16)

# Bottom Row: Comparison Bar Chart
ax3 = plt.subplot2grid((2, 2), (1, 0), colspan=2)
df_melted = df_report.melt(id_vars='Class', value_vars=['TRAIN', 'VAL', 'TEST'], var_name='Split', value_name='Count')
sns.barplot(data=df_melted, x='Class', y='Count', hue='Split', ax=ax3)
ax3.set_title("Split Comparison (Train vs Val vs Test)", fontsize=16)
ax3.tick_params(axis='x', rotation=30)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
import os
import shutil
from tqdm.auto import tqdm
from pathlib import Path
from concurrent.futures import ThreadPoolExecutor, as_completed

# 1. Setup Folders
new_dataset_path = Path('/content/doclaynet_8k_raw')
if new_dataset_path.exists():
    shutil.rmtree(new_dataset_path)

for s in ['train', 'val', 'test']:
    os.makedirs(new_dataset_path / f'images/{s}', exist_ok=True)
    os.makedirs(new_dataset_path / f'labels/{s}', exist_ok=True)

# 2. Set Targets (Total 8,000)
targets = {'train': 6864, 'val': 640, 'test': 496}

# 3. Fast Worker Function
def save_data_worker(img, bboxes, cat_ids, split, f_name):
    try:
        img_file = new_dataset_path / f'images/{split}' / f"{f_name}.png"
        lbl_file = new_dataset_path / f'labels/{split}' / f"{f_name}.txt"
        img.save(img_file)

        img_w, img_h = img.size
        with open(lbl_file, 'w') as f:
            for bbox, cat_id in zip(bboxes, cat_ids):
                x, y, w, h = bbox
                # YOLO format (class-1 for 0-indexing)
                f.write(f"{cat_id-1} {(x+w/2)/img_w:.6f} {(y+h/2)/img_h:.6f} {w/img_w:.6f} {h/img_h:.6f}\n")
        return True
    except Exception as e:
        return f"Error: {e}"

# 4. Multi-Threaded Loop (No category check)
print(f"⚡ Starting Raw Extraction to: {new_dataset_path}")

for split in ['train', 'val', 'test']:
    target_limit = targets[split]
    batch_size = 100 # Increased batch size for even more speed

    print(f"\n--- Extracting {split.upper()} ---")
    pbar = tqdm(total=target_limit, desc=f"Progress")

    with ThreadPoolExecutor(max_workers=8) as executor: # Increased workers
        futures = []

        # We take exactly the first N rows using slicing [:target_limit]
        for idx, row in enumerate(dataset[split].select(range(target_limit))):
            meta = row.get('metadata', {})

            # Safe naming fallback
            doc_id = meta.get('doc_hash') or meta.get('doc_filename') or f"raw_{idx}"
            page_num = meta.get('page_no') if meta.get('page_no') is not None else f"p{idx}"
            f_name = f"{split}_{doc_id}_{page_num}"

            futures.append(executor.submit(
                save_data_worker, row['image'], row['bboxes'], row['category_id'], split, f_name
            ))

            if len(futures) >= batch_size:
                for future in as_completed(futures):
                    future.result()
                    pbar.update(1)
                futures = []

        for future in as_completed(futures):
            future.result()
            pbar.update(1)

print("\n✅ DONE! 8,000 images extracted without metadata filtering.")
```

```python
import os
from tqdm.auto import tqdm
from pathlib import Path
from concurrent.futures import ThreadPoolExecutor, as_completed

# 1. Setup Folders
output_path = Path('/content/doclaynet_ratio_mapped_8k_yolo')
for s in ['train', 'val', 'test']:
    os.makedirs(output_path / f'images/{s}', exist_ok=True)
    os.makedirs(output_path / f'labels/{s}', exist_ok=True)

print("Starting LOCAL Stratified Extraction from your downloaded v1.1 data...")

# 2. EXACT Category Quotas
category_targets = {
    'train': {
        'financial_reports': 2183, 'manuals': 1428, 'scientific_articles': 1167,
        'laws_and_regulations': 1125, 'patents': 542, 'government_tenders': 419
    },
    'validation': { # HF datasets usually use 'validation' instead of 'val'
        'financial_reports': 203, 'manuals': 133, 'scientific_articles': 109,
        'laws_and_regulations': 105, 'patents': 51, 'government_tenders': 39
    },
    'test': {
        'financial_reports': 158, 'manuals': 103, 'scientific_articles': 84,
        'laws_and_regulations': 82, 'patents': 39, 'government_tenders': 30
    }
}

split_map = {'train': 'train', 'validation': 'val', 'test': 'test'}

# 3. The Shielded Worker
def save_file_worker(img, bboxes, cat_ids, split, f_name):
    img_file = output_path / f'images/{split}' / f"{f_name}.png"
    lbl_file = output_path / f'labels/{split}' / f"{f_name}.txt"

    if os.path.exists(img_file) and os.path.exists(lbl_file):
        return True

    try:
        img.save(img_file)
        img_w, img_h = img.size
        with open(lbl_file, 'w') as f:
            for bbox, cat_id in zip(bboxes, cat_ids):
                x, y, w, h = bbox

                # YOLO format with float guards
                center_x = min(1.0, max(0.0, (x + w / 2) / img_w))
                center_y = min(1.0, max(0.0, (y + h / 2) / img_h))
                norm_w = min(1.0, max(0.0, w / img_w))
                norm_h = min(1.0, max(0.0, h / img_h))

                yolo_cat = cat_id - 1
                f.write(f"{yolo_cat} {center_x:.6f} {center_y:.6f} {norm_w:.6f} {norm_h:.6f}\n")
        return True
    except Exception as e:
        return f"Error on {f_name}: {e}"

# 4. Batched Execution Logic (Using your ALREADY DOWNLOADED dataset variable)
for hf_split, yolo_split in split_map.items():
    targets = category_targets[hf_split]
    total_target = sum(targets.values())

    collected = {cat: 0 for cat in targets.keys()}
    total_collected = 0
    batch_size = 75

    print(f"\n--- Processing {yolo_split.upper()} Split ---")
    scan_pbar = tqdm(total=total_target, desc=f"Scanning", position=0, leave=True, dynamic_ncols=True)
    save_pbar = tqdm(total=total_target, desc=f"Saving  ", position=1, leave=True, dynamic_ncols=True)

    with ThreadPoolExecutor(max_workers=6) as executor:
        futures = []

        # Iterating over the local dataset you already loaded
        for row in dataset[hf_split]:
            if total_collected >= total_target:
                break

            doc_category = row['metadata']['doc_category']

            if collected[doc_category] >= targets[doc_category]:
                continue

            img = row['image']
            bboxes = row['bboxes']
            cat_ids = row['category_id']

            f_name = f"doc_{yolo_split}_{doc_category}_{total_collected}"
            futures.append(executor.submit(save_file_worker, img, bboxes, cat_ids, yolo_split, f_name))

            collected[doc_category] += 1
            total_collected += 1
            scan_pbar.update(1)

            if len(futures) >= batch_size:
                for future in as_completed(futures):
                    result = future.result()
                    if result is not True: tqdm.write(f"{result}")
                    save_pbar.update(1)
                futures = []

        for future in as_completed(futures):
            result = future.result()
            if result is not True: tqdm.write(f"{result}")
            save_pbar.update(1)

    scan_pbar.close()
    save_pbar.close()
    print(f"{yolo_split.upper()} Quotas filled: {collected}")

print("\nEXTRACTION COMPLETE!")

# Final Verification Count
print("\n--- FINAL FILE COUNTS ---")
for s in ['train', 'val', 'test']:
    img_count = len(os.listdir(output_path / f'images/{s}'))
    lbl_count = len(os.listdir(output_path / f'labels/{s}'))
    print(f"{s.capitalize()}: {img_count} Images, {lbl_count} Labels")
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuration
labels_path = '/content/doclaynet_ratio_mapped_8k_yolo/labels'
class_names = [
    "Caption", "Footnote", "Formula", "List-item", "Page-footer",
    "Page-header", "Picture", "Section-header", "Table", "Text", "Title"
]

def get_label_stats(split_name=None):
    stats = {name: 0 for name in class_names}

    # Determine which folders to scan
    folders = [split_name] if split_name else ['train', 'val', 'test']

    for folder in folders:
        path = os.path.join(labels_path, folder)
        if not os.path.exists(path): continue
        for label_file in os.listdir(path):
            with open(os.path.join(path, label_file), 'r') as f:
                for line in f:
                    try:
                        class_id = int(line.split()[0])
                        stats[class_names[class_id]] += 1
                    except: continue
    return stats

# --- 1. COLLECT ALL DATA ---
all_stats = get_label_stats()  # Total
train_stats = get_label_stats('train')
val_stats = get_label_stats('val')
test_stats = get_label_stats('test')

# Create a combined DataFrame for the report
df_report = pd.DataFrame({
    'Class': class_names,
    'TRAIN': [train_stats[c] for c in class_names],
    'VAL': [val_stats[c] for c in class_names],
    'TEST': [test_stats[c] for c in class_names],
    'TOTAL': [all_stats[c] for c in class_names]
})

# --- 2. PRINT TEXT SUMMARY ---
print(" --- DETAILED LABEL COUNT REPORT --- ")
print(df_report.to_string(index=False))
print(f"\n GRAND TOTAL BOXES: {df_report['TOTAL'].sum():,}")

# --- 3. VISUALIZATION ---
fig = plt.figure(figsize=(20, 15))
plt.suptitle("ArXiv Dataset Label Distribution Analysis", fontsize=22, fontweight='bold')

# Top Row: Grand Total Charts
ax1 = plt.subplot2grid((2, 2), (0, 0))
ax2 = plt.subplot2grid((2, 2), (0, 1))

# Bar Chart (Total)
sns.barplot(x='TOTAL', y='Class', data=df_report.sort_values('TOTAL', ascending=False), ax=ax1, palette='viridis')
ax1.set_title("Grand Total Counts (All Splits)", fontsize=16)

# Pie Chart (Total %)
ax2.pie(df_report['TOTAL'], labels=df_report['Class'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', 11))
ax2.set_title("Overall Percentage Breakdown", fontsize=16)

# Bottom Row: Comparison Bar Chart
ax3 = plt.subplot2grid((2, 2), (1, 0), colspan=2)
df_melted = df_report.melt(id_vars='Class', value_vars=['TRAIN', 'VAL', 'TEST'], var_name='Split', value_name='Count')
sns.barplot(data=df_melted, x='Class', y='Count', hue='Split', ax=ax3)
ax3.set_title("Split Comparison (Train vs Val vs Test)", fontsize=16)
ax3.tick_params(axis='x', rotation=30)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
